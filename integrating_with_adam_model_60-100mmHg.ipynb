```python
def deltaR1_along_capillary_tissue(pO2_after_list,pO2_before_list,pO2Tissueafter,pO2Tissuebefore,Safterlist,Sbeforelist,x_axis,bloodvolume,title):
    # B0=1.5 # Tesla
    Hct=0.47 # from 0-1
    Hb=5.15 # mmol Hb tetramer/L plasma
    fwatere=(0.7*Hct)/((0.7*Hct)+0.95*(1-Hct))
    #print(fwatere)
    r1deoxyHb=0.033 #s-1 L plasma in erythrocyte / mmol Hb tetramer
    r1=0.00025 #s-1/mmHg


    delta_R1voxeltotal_list=np.zeros(len(pO2_before_list[0]))

    fig1 = plt.figure()
    ax1 = fig1.add_subplot(111)

    for j in range(0,len(pO2_before_list)):
        for i in range(0,len(pO2_before_list[0])):
            #R1arterialbefore:
            pO2before=pO2_before_list[j,i] #mmHg
            Sbefore=((23400*((pO2before**3)+150*pO2before)**(-1)+1)**(-1))

            #R1arterialafter:
            pO2after=pO2_after_list[j,i] #mmHg
            Safter=((23400*((pO2after**3)+150*pO2after)**(-1)+1)**(-1))

            deltapO2=pO2after-pO2before
            deltapO2_tissue=pO2Tissueafter-pO2Tissuebefore
            #deltaR1plasma=R1plasmabefore-R1plasmaafter
            deltaR1plasma=r1*deltapO2

            deltaR1bloodtotal=fwatere*r1deoxyHb*Hb*((1-Safter)-(1-Sbefore))+(1-fwatere)*(deltaR1plasma)
            #print('delta R1 blood total is '+str(deltaR1bloodtotal))

            deltaR1_voxeltotal=bloodvolume*deltaR1bloodtotal+(1-bloodvolume)*r1*deltapO2_tissue
            delta_R1voxeltotal_list[i]=deltaR1_voxeltotal

        plt.plot(x_axis,delta_R1voxeltotal_list,label=str(pO2after))


    colormap = plt.cm.cool #nipy_spectral, Set1,Paired   
    colors = [colormap(i) for i in np.linspace(0.1, 1,len(ax1.lines))]
    for i,j in enumerate(ax1.lines):
        j.set_color(colors[i])

    plt.plot(x_axis,np.zeros(len(x_axis)),'k--')

    plt.title(str(title))
    plt.xlabel('X along capillary')
    plt.ylabel('$\Delta R1$ (s-1)')
    plt.legend(labels=[60,65,70,75,80,85,90,95,100],bbox_to_anchor=(1, 1))
    plt.ylim(-0.0075,0.0075)
    plt.xlim(0,600)
```
