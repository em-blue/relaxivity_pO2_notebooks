```python


def deltapO2_vs_deltaR1_fullscale(pO2_after_list,pO2_before_list,Sbeforelist,Safterlist,x_axis,bloodvolume,title):
    # B0=1.5 # Tesla
    Hct=0.47 # from 0-1
    Hb=5.15 # mmol Hb tetramer/L plasma
    fwatere=(0.7*Hct)/((0.7*Hct)+0.95*(1-Hct))
    #print(fwatere)
    r1deoxyHb=0.033 #s-1 L plasma in erythrocyte / mmol Hb tetramer
    r1=0.00025 #s-1/mmHg
    
    delta_R1voxeltotal_list=np.zeros((len(pO2_before_list)))
    deltapO2_list=np.zeros((len(pO2_before_list)))
    #deltaTissuepO2_list=np.zeros((len(pO2_before_list)))
    
    fig2 = plt.figure()
    ax2 = fig2.add_subplot(111)

    for i in range(0,len(pO2_before_list)):
        #R1arterialbefore:
        pO2before=pO2_before_list[i] #mmHg
        #pO2Tissuebefore=pO2Tissuebefore_list[i]
        #Sbefore=Sbeforelist[i]
        Sbefore=((23400*((pO2before**3)+150*pO2before)**(-1)+1)**(-1))

        #R1arterialafter:
        pO2after=pO2_after_list[i] #mmHg
        #pO2Tissueafter=pO2Tissueafter_list[i]
        Safter=((23400*((pO2after**3)+150*pO2after)**(-1)+1)**(-1))
        #Safter=Safterlist[i]

        deltapO2=pO2after-pO2before
        deltapO2_list[i]=deltapO2
        #print(deltapO2_list)
        #deltaR1plasma=R1plasmabefore-R1plasmaafter
        deltaR1plasma=r1*deltapO2

        deltaR1bloodtotal=fwatere*r1deoxyHb*Hb*((1-Safter)-(1-Sbefore))+(1-fwatere)*(deltaR1plasma)
        #print('delta R1 blood total is '+str(deltaR1bloodtotal))

       # deltaTissuepO2=pO2Tissueafter-pO2Tissuebefore
        #deltaTissuepO2_list[i]=deltaTissuepO2

        #deltaR1_voxeltotal=bloodvolume*deltaR1bloodtotal+(1-bloodvolume)*r1*deltaTissuepO2
       # delta_R1voxeltotal_list[i]=deltaR1_voxeltotal
        delta_R1voxeltotal_list[i]=deltaR1bloodtotal

    plt.plot(x_axis,delta_R1voxeltotal_list,'c')
    plt.plot(x_axis,np.zeros(len(x_axis)),'k--')

    #print(deltapO2_list)
    #print(delta_R1voxeltotal_list)

    plt.title(str(title))
    plt.xlabel('X along capillary')
    plt.ylabel('$\Delta R1$ (s-1)')
    plt.ylim(-0.02,0.005)
    plt.xlim(0,600)
    
```

```python
def deltaR1_along_capillary_tissue(pO2_after_list,pO2_before_list,deltatissuepO2,x_axis,bloodvolume,title):
    # B0=1.5 # Tesla
    Hct=0.47 # from 0-1
    Hb=5.15 # mmol Hb tetramer/L plasma
    fwatere=(0.7*Hct)/((0.7*Hct)+0.95*(1-Hct))
    #print(fwatere)
    r1deoxyHb=0.033 #s-1 L plasma in erythrocyte / mmol Hb tetramer
    r1=0.00025 #s-1/mmHg
    
    delta_R1voxeltotal_list=np.zeros((len(pO2_before_list)))
    deltapO2_list=np.zeros((len(pO2_before_list)))
    #deltaTissuepO2_list=np.zeros((len(pO2_before_list)))
    
    fig2 = plt.figure()
    ax2 = fig2.add_subplot(111)

    for i in range(0,len(pO2_before_list)):
        #R1arterialbefore:
        pO2before=pO2_before_list[i] #mmHg
        #pO2Tissuebefore=pO2Tissuebefore_list[i]
        #Sbefore=Sbeforelist[i]
        Sbefore=((23400*((pO2before**3)+150*pO2before)**(-1)+1)**(-1))

        #R1arterialafter:
        pO2after=pO2_after_list[i] #mmHg
        #pO2Tissueafter=pO2Tissueafter_list[i]
        Safter=((23400*((pO2after**3)+150*pO2after)**(-1)+1)**(-1))
        #Safter=Safterlist[i]

        deltapO2=pO2after-pO2before
        deltapO2_list[i]=deltapO2
        #print(deltapO2_list)
        #deltaR1plasma=R1plasmabefore-R1plasmaafter
        deltaR1plasma=r1*deltapO2

        deltaR1bloodtotal=fwatere*r1deoxyHb*Hb*((1-Safter)-(1-Sbefore))+(1-fwatere)*(deltaR1plasma)
        #print('delta R1 blood total is '+str(deltaR1bloodtotal))

       # deltaTissuepO2=pO2Tissueafter-pO2Tissuebefore
        #deltaTissuepO2_list[i]=deltaTissuepO2

        deltaR1_voxeltotal=bloodvolume*deltaR1bloodtotal+(1-bloodvolume)*r1*deltatissuepO2
        delta_R1voxeltotal_list[i]=deltaR1_voxeltotal
        # delta_R1voxeltotal_list[i]=deltaR1bloodtotal

    plt.plot(x_axis,delta_R1voxeltotal_list,'c')
    plt.plot(x_axis,np.zeros(len(x_axis)),'k--')

    #print(deltapO2_list)
    #print(delta_R1voxeltotal_list)

    plt.title(str(title))
    plt.xlabel('X along capillary')
    plt.ylabel('$\Delta R1$ (s-1)')
    plt.ylim(-0.005,0.005)
    plt.xlim(0,600)

```
